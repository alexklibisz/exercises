# Logistic Regression

Used for binary classification. LR training is typically implemented using maximum-likelihood estimation, but can also be done using gradient descent which is simpler.

### Logistic Function

Use the logistic function to evaluate a value as a probability:

$g(z) = \frac{1}{1 + e^{-z}}$

### Cost Function

The cost function for a single example:

$\text{Cost}(h_{\theta}(x^{(i)}), y^{(i)}) = -y \cdot \text{log}(h_\theta(x) - (1 - y) \cdot \text{log}(1 - h_\theta(x))$

Where $y_i = \{0,1\}$

The complete cost function for all examples is:

$J(\theta) = \frac{-1}{m} \sum_{i=1}^m \text{Cost}(h_{\theta}(x^{(i)}), y^{(i)}) = \frac{-1}{m} \sum_{i=1}^m -y^{(i)} \cdot \text{log}(h_\theta(x^{(i)}) - (1 - y^{(i)}) \cdot \text{log}(1 - h_\theta(x^{(i)}))$

### Cost Function Derivative (for gradient descent)

The derivative of the cost function, to be used with gradient descent, is:

$\frac{\partial}{\partial} \theta_j J(\theta) = \frac{1}{m} \sum_{i=1}^m (h_\theta(x^{(i)}) - y^{(i)}) \cdot x_j^{(i)}$ 

### Gradient Descent Update

(Same as linear regression)



```python
# Just plotting the logistic function.
import matplotlib.pyplot as plt
from math import exp

def logistic(n):
    return 1 / (1 + exp(-n))

x = list(range(-10,11))
y = list(map(logistic, x))

plt.title('Logistic Function -10 to 10')
plt.plot(x,y)
plt.ylim(-0.2,1.2)
plt.show()
```

## Inference using Logistic Regression

Similar to Linear Regression, we use the linear combination of the features and their weights. The resulting value is passed to the logistic function, which maps the value somewhere between 0 and 1. If it's closer to 1, it belongs to class 1. If it's closer to 0, it belongs to class 0.

```python
# Logistic Function
import matplotlib.pyplot as plt
import random
from math import exp

def logistic(n):
    return 1 / (1 + exp(-n))
 
# Initialize X and shuffle it.
X = list(range(-5,6))
random.shuffle(X)

# Initialize a theta of random weights.
theta = list(map(lambda x: random.uniform(-1,2), x))

# Apply the weights with a bias term 1 to the X's to get the Y.
# This gives you: Y[0] = 1 + theta[0] * X[0], ...
Y = list(map(lambda p: logistic(1 + p[0]*p[1]), list(zip(theta, X))))
   
plt.title('Logistic Function -10 to 10')
plt.plot(X,Y,'r.')
plt.plot(X,[0.5]*len(X),'b')
plt.ylim(-0.2,1.2)
plt.show()
```

## Scikit-Learn Logistic Regression Implementation on a Dataset

Using the popular [Pima Indians Diabetes Dataset](https://archive.ics.uci.edu/ml/datasets/Pima+Indians+Diabetes) to predict whether the person has diabetes.



```python
import pandas as pd
import matplotlib.pyplot as plt
from math import floor
from sklearn.metrics import f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

names = ['times_pregnant', 'plasma_glucose_concentration', 'diastolic_blood_pressure', 'triceps_skin_fold_thickness', 'serum_insulin', 'BMI', 'diabetes_pedigree_function', 'age', 'diabetes']
df = pd.read_csv('./data/pima-indians-diabetes.data', names=names, header=None)

X = df[['times_pregnant', 'plasma_glucose_concentration', 'BMI', 'age']].values
Y = df['diabetes'].values

# Scale all features in X.
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Split for training and testing.
size = floor(X.shape[0] * 0.7)
X_train = X[0:size]
Y_train = Y[0:size]
X_test = X[size:]
Y_test = Y[size:]

# Fit and evaluate the probabilities for the model.
model = LogisticRegression()
model.fit(X_train,Y_train)
predicted = model.predict_proba(X_test)
binary = model.predict(X_test)
fscore = f1_score(Y_test, binary)

plt.title('Test Set, size %d, fscore = %lf' % (len(predicted), fscore))
plt.plot(predicted, 'r.')
plt.plot([0.5]*len(predicted), 'b-')
plt.show()
```

# My Logistic Regression Implementation on a Dataset

Using the same dataset as the scikit-learn example above. Implemented as a class to emulate the scikit-learn interface. Using the objective function taught on the Coursera machine learning course.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import floor, exp
from sklearn.metrics import f1_score
from sklearn.preprocessing import StandardScaler


'''
LR Class Implementation
'''
class LogisticRegression():
    def __init__(self):
        # The theta vector is maintained to 
        # go from fitting to evaluating.
        self._theta = None
        
    def fit(self, X, Y, alpha=0.0005, numIter=1000, verbose=False):
        '''Gradient descent used to fit a theta vector to X.
        '''
        
        # Extract the shape from (m by n) matrix X.
        m,n = X.shape
        
        # X and theta with a column of ones.
        X = np.column_stack((np.ones(shape=(m,1)), X))
        self._theta = np.ones(n + 1)
        
        # Used for getting gradient.
        XT = X.transpose()
        
        # Theta update loop with gradient descent used to update weights.
        for i in range(0, numIter):
            hypothesis = self.predict_proba(X,fixX=False)
            error = self._cost(hypothesis,Y)
            
            if verbose and i % 100 is 0:
                print('iteration %d mean error %lf' % (i, sum(error) / len(error)))
            
            slope = (1/m) * np.dot(XT, error)
            change = alpha * slope
            self._theta = self._theta - change
        
        return self._theta
        
    def predict_proba(self, X, fixX=True):
        
        # Hack to prepend a column of ones if necessary.
        if fixX:
            X = np.column_stack((np.ones(shape=(X.shape[0],1)), X))
            
        # Multiply X by the theta weights to get an (m x 1) column vector.
        weighted = np.dot(X, self._theta)
        
        # Evaluate each of the weighted values using the logistic function.
        return np.array([self._logistic(w) for w in weighted])
    
    def predict(self, X):
        proba = self.predict_proba(X)
        return np.array([round(p) for p in proba])
    
    def _cost(self,h,Y):
        '''Objective function introduced in coursera ML course
        Vectorized implementation described here:
        https://www.coursera.org/learn/machine-learning/discussions/weeks/3/threads/0DKoqvTgEeS16yIACyoj1Q'''
        m = h.shape[0]
        t1 = np.multiply(Y, np.log(h))
        t2 = np.multiply(1 - Y, np.log(1 - h))
        return (-1 / m) * (t1 - t2)
    
    def _logistic(self, n):
        return 1 / (1 + exp(-n))
    
'''
Read and evaluate dataset using above implementation
'''
names = ['times_pregnant', 'plasma_glucose_concentration', 'diastolic_blood_pressure', 'triceps_skin_fold_thickness', 'serum_insulin', 'BMI', 'diabetes_pedigree_function', 'age', 'diabetes']
df = pd.read_csv('./data/pima-indians-diabetes.data', names=names, header=None)

X = df[['times_pregnant', 'plasma_glucose_concentration', 'BMI', 'age']].values
Y = df['diabetes'].values

# Scale all features in X.
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Split for training and testing.
size = floor(X.shape[0] * 0.7)
X_train = X[0:size]
Y_train = Y[0:size]
X_test = X[size:]
Y_test = Y[size:]

model = LogisticRegression()
model.fit(X_train,Y_train,verbose=False, numIter=1000)

predicted = model.predict_proba(X_test)
binary = model.predict(X_test)
fscore = f1_score(Y_test, binary)

plt.title('Test Set, size %d, fscore %lf' % (len(predicted), fscore))
plt.ylim(-0.2, 1.2)
plt.plot(predicted, 'r.')
plt.plot([0.5]*len(predicted), 'b-')
plt.show()
```
