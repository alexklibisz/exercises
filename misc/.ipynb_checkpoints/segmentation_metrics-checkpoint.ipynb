```python
# Plotting various segmentation metrics from to get a more intuitive understanding for how they work.
import tifffile as tif
import numpy as np
import matplotlib.pyplot as plt

eps = 1e-7

def prec(t, p):
    p = p.round()
    return np.sum(t * p) / (np.sum(p) + eps)

def reca(t, p):
    p = p.round()
    return np.sum(t * p) / (np.sum(t) + eps)

def F1(t, p):
    pr = prec(t, p)
    re = reca(t, p)
    return (2 * pr * re) / (pr + re + eps)

def jacc(t, p):
    return jacc_soft(t, p.round())

def jacc_soft(t, p):
    I = np.sum(t * p)
    U = np.sum(t) + np.sum(p) - I
    return I / (U + eps)

def dice(t, p):
    return dice_soft(t, p.round())
    
def dice_soft(t, p):
    I = np.sum(t * p)
    return (2. * I) / (np.sum(t) + np.sum(p) + 1e-7)

def dicesq(t, p):
    n = 2 * np.sum(t * p)
    d = np.sum(t**2) + np.sum(p**2)
    return n / (d + eps)

def logloss(t, p):
    ll = (t * np.log(p + eps)) + (1 - t) * np.log(1 - p + eps)
    return -np.mean(ll)

# Double check metric implementations against scipy implementations.
from sklearn.metrics import f1_score, precision_score, recall_score, log_loss
from scipy.spatial.distance import dice as _dice
from scipy.spatial.distance import jaccard as _jacc
np.random.seed(10)
for _ in range(1000):
    t = np.random.rand(100,).round()
    p = np.random.rand(100,).round()
    assert abs(F1(t,p) - f1_score(t,p)) < 1e-5
    assert abs(prec(t,p) - precision_score(t,p)) < 1e-5
    assert abs(reca(t,p) - recall_score(t,p)) < 1e-5
    assert abs(dice(t,p) - (1 - _dice(t,p))) < 1e-5
    assert abs(jacc(t,p) - (1 - _jacc(t,p))) < 1e-5
    
print('Done')
```

```python
# Use images from the 8x8 digits dataset to evaluate all of the metrics.
from sklearn.datasets import load_digits
digits = load_digits().images / 16.

np.random.seed(111)
idxs = [np.random.randint(0,200) for _ in range(2000)]
samples = digits[idxs]
fig,_ = plt.subplots(10,10)
for ix, ax in enumerate(fig.axes):
    ax.imshow(samples[ix], cmap='gray')
    ax.axis('off')
plt.show()
```

```python
ind_funcs = [F1, prec, reca]
dep_funcs = [logloss, jacc, jacc_soft, dice, dice_soft, dicesq]

fig, axes = plt.subplots(len(ind_funcs), len(dep_funcs), figsize=(20, 15))

for iy, indf in enumerate(ind_funcs):
    for ix, depf in enumerate(dep_funcs):    
        ax = fig.axes[iy * len(dep_funcs) + ix]
        ax.set_ylabel(depf.__name__)
        ax.set_xlabel(indf.__name__)
        ax.set_xlim(0,1)
        ax.set_ylim(0,2.5)
        ax.yaxis.grid(color='gray', linestyle='dashed')
        ax.xaxis.grid(color='gray', linestyle='dashed')
        x = [indf(samples[i].round(), samples[(i+1)%len(samples)]) for i in range(len(samples))]
        y = [depf(samples[i].round(), samples[(i+1)%len(samples)]) for i in range(len(samples))]
        ax.scatter(x,y)
        
plt.subplots_adjust(left=None, wspace=0.4, hspace=0.4, right=None)
plt.show()
```
